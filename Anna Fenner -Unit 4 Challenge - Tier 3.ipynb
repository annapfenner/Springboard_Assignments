#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#transposing data so each row track is one borough over time 
properties_df = properties.transpose().reset_index()
properties_df
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#set column names to be the frist row in data
properties_df.rename(columns = properties_df.iloc[0], inplace = True)
properties_df.head()
```

```python
#drop the first row in data as used it to rename the columns 
properties_data= properties_df.drop(properties.index[0])

properties_data.head()
```

```python
#renaming a few columns for the London borough and ID
properties_data.rename(columns = {'Unnamed: 0': 'London borough', pd.NaT : 'ID'}, inplace = True)
properties_data.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#restructuring the data by using melt so keeping the borough and ID columns then the other columns became two columns: date and avg price
transform_data = pd.melt(properties_data, id_vars =['London borough', 'ID'])
transform_data.head()
```

```python
#rename the new columns to 'Date' and 'Avg_Price'
transform_data = transform_data.rename(columns ={'variable': 'Date', 'value': 'Avg_Price'})
transform_data.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Checking to see if column 'Avg_Price' is a float and then converting column to float
transform_data['Avg_Price'].dtypes
transform_data['Avg_Price']=transform_data['Avg_Price'].astype(float)
transform_data['Avg_Price'].dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#plotting missing data, determing how many missing data and overall shape of dataframe for comparison
transform_data.isna().sum().plot(kind='bar')
transform_data.isna().any().sum()
transform_data.shape
```

```python
#drop NA values 
dropna_data= transform_data.dropna()
```

```python
#plot missing values
print(dropna_data.isna().sum().any())
dropna_data.isna().sum().plot(kind='bar')
```

```python
#determining number of boroughs in dataframe
print(dropna_data['London borough'].unique())
print(dropna_data['London borough'].nunique())
print(dropna_data['London borough'].dtypes)
```

```python
#removing the data that are not boroughs
not_boroughs = ['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England']

notboroughs_rows = dropna_data['London borough'].isin(not_boroughs)
borough_data=dropna_data[~notboroughs_rows]

#checking to make sure only ones remaining are the actual boroughs
print(borough_data['London borough'].unique())
print(borough_data['London borough'].nunique())
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
#visual data to see what one borough looks like over 20 years
harrow_data = borough_data[borough_data['London borough']=='Harrow']
harrow_data.plot(kind='line', x='Date', y='Avg_Price')
plt.xlabel('Years')
plt.ylabel('Avg_Price')
plt.title("Harrow Prices over the Years")
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
#creating a year column in dataframe so eventually will take the mean of the avg_price for each year for each borough
borough_data['Year'] = borough_data['Date'].apply(lambda x: x.year)
borough_data.head()
```

```python
#find average price for each borough for each year
borough_avg = borough_data.groupby(["London borough", "Year"])["Avg_Price"].mean()
print(borough_avg.head())
print(borough_avg.shape)

#resetting index as London borough and Year are part of the index
borough_df= borough_avg.reset_index()

#check to make sure the reset index worked
print(borough_df.head())
borough_df.shape
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#creating a function to return a ratio of avg price for 2018 to avg price for 1998

def create_price_ratio(x) :
    year_1998 = float(x['Avg_Price'][x['Year']==1998])
    year_2018 = float(x['Avg_Price'][x['Year']==2018])
    ratio = [year_2018/year_1998]
    return ratio

#check to see if works if call on one borough 
create_price_ratio(borough_df[borough_df['London borough']=='Harrow'])
```

```python
#create an empty dictionary
dict_ratio = {}

#create a for loop to go through each borough and call the function create_price_ratio
for y in borough_df['London borough'].unique():
    dict_ratio[y]=create_price_ratio(borough_df[borough_df['London borough']==y])
print(dict_ratio)
```

```python
#need to make dictionary a dataframe and print out dataframe and shape
df_ratio=pd.DataFrame.from_dict(dict_ratio)
print(df_ratio)
print(df_ratio.shape)
```

```python
#transpose dataframe
df =df_ratio.T
df
```

```python
#reset index again so borough is a column and not index
df_reset_index= df.reset_index()

#rename columns
df_rename_columns = df_reset_index.rename(columns={'index': 'Borough', 0: 'Ratio'})
print(df_rename_columns)
```

```python
#sort dataframe so highest ratio is on top
df_sorted = df_rename_columns.sort_values(by=['Ratio'], ascending=False)
print(df_sorted)
```

```python
#creating a top 10 dataframe to then use for a boxplot
top10 = df_sorted[:10]
```

```python
#make a horizontal bar chart to see what boroughs have the highest ratio
ax= top10[['Borough', 'Ratio']].plot(kind='barh')
ax.set_yticklabels(top10.Borough)
```

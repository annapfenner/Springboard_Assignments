```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
  

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def select_blah_blah(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT (m1.firstname || ' ' || m1.surname) AS first_last_name, (m2.firstname || ' ' || m1.surname) AS member_who_recommended
        FROM Members AS m1
        LEFT JOIN Members AS m2
        ON m1.memid = m2.recommendedby
        ORDER BY first_last_name ASC;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"

    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
        select_blah_blah(conn)
 
 
if __name__ == '__main__':
    main()
    
```

```python
#question 10
def select_question_10(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
       SELECT name, SUM(facility_revenue) AS total_facility_revenue FROM (SELECT name, CASE WHEN memid = 0 THEN guestcost * slots ELSE membercost * slots END AS facility_revenue FROM Bookings as b JOIN Facilities as f USING (facid)) AS sub GROUP BY name ORDER BY total_facility_revenue LIMIT 3;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)



```

```python
#question 12
def select_question_12(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
       SELECT facid, COUNT(memid) 
       FROM Bookings
    WHERE memid !=0
    GROUP BY facid;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
#question 13
def select_question_13(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
       SELECT sub.months, COUNT(sub.memid) 
       FROM (SELECT MONTH( starttime ) AS months, memid FROM Bookings WHERE memid !=0) AS sub GROUP BY sub.months;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)



```
